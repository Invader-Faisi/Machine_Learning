# Univariant Analysis

```python
import pandas as pd
import seaborn as sns
data = pd.read_csv('train.csv')
data.head()
```

## Categorical Data

## a. Countplot

```python
# survived, pclass, sex, sibsp, parch, cabin, embarked are categorical cols
sns.countplot(x=data['Survived'])
#data['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x=data['Pclass'])
```

```python
sns.countplot(x=data['Sex'])
```

```python
sns.countplot(x=data['Embarked'])
```

## b. Pie Char

```python
data['Survived'].value_counts().plot(kind = 'pie', autopct = '%0.2f')
```

## Numerical Data

## a. Histogram

```python
# to see the distribution of numerical data use histogram
import matplotlib.pyplot as plt
plt.hist(data['Age'], bins = 50)
```

## Distplot

```python
sns.displot(data['Age'])
```

```python
sns.distplot(data['Age'])
# this shows the probability of any number
```

## c. Boxplot

```python
sns.boxplot(data['Fare'])
```

```python
# skewness
data['Age'].skew()
```
